```python
# Missing Values Check
print(df.isnull().sum())

# Remove Duplicates
df = df.drop_duplicates()

# Standardize Column Names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Check Data Types
print(df.dtypes)

# Save Cleaned Dataset
df.to_csv(r"C:\Users\kirut\Downloads\cleaned_supermarket_sales.csv", index=False)

print("Data Cleaning Completed Successfully")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df["sales"])
plt.title("Outlier Detection - Sales")
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["sales"], bins=20)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
df["product_line"].value_counts().plot(kind="bar")
plt.title("Product Line Count")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="quantity", y="sales", data=df)
plt.title("Quantity vs Sales")
plt.show()
```
